```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

from astropy.io import fits
from astropy.convolution import Gaussian2DKernel, convolve

from scipy.ndimage import median_filter
```

```python
base_dir = '/Users/acarter/Documents/DIRECT_IMAGING/CORONAGRAPHY_PIPELINE/20220719_F1140/BACKUP/IMGPROCESS/'
file = base_dir + 'jw01386004001_04101_00001_mirimage_calints.fits'
```

```python
hdu = fits.open(file)
dq = hdu['DQ'].data
sci = hdu['SCI'].data

# mask = np.where(dq != 0)
# sci[mask] = 999
```

```python
plt.figure(figsize=[12,15])
plt.imshow(sci[0], norm=LogNorm())
```

```python
kernel = Gaussian2DKernel(2)
conv = convolve(sci[0], kernel)
```

```python
plt.figure(figsize=[12,15])
plt.imshow(conv, norm=LogNorm())
```

```python
diff = sci[0] - conv
```

```python
plt.figure(figsize=[12,15])
plt.imshow(diff)
```

```python
test = sci[0,:,15:225]

blurred = median_filter(test, size=3)
diff = np.subtract(test,blurred)
threshold = 10*np.std(diff)

hot_pixels = np.nonzero((np.abs(diff[1:-1,1:-1])>threshold) )
hot_pixels = np.array(hot_pixels) + 1

fixed_image = np.copy(test) #This is the image with the hot pixels removed
for y,x in zip(hot_pixels[0],hot_pixels[1]):
    fixed_image[y,x]= blurred[y,x]
```

```python
plt.figure(figsize=[12,15])
plt.imshow(test, norm=LogNorm())    
plt.figure(figsize=[12,15])
plt.imshow(blurred, norm=LogNorm())
plt.figure(figsize=[12,15])
plt.imshow(fixed_image, norm=LogNorm())
```
